# Reading data

```python
import pandas as pd
import numpy as np 
import pickle
```

```python
infile = open('Full_Cortex_Dataset.pkl','rb')
DF1 = pickle.load(infile)
infile.close()
```

```python
infile = open('Full_Read_Dataset.pkl','rb')
DFread = pickle.load(infile)
infile.close()
```

# Section 3: Double Lasso on Top 10% Engagement variables

```python
from sklearn.linear_model import LassoCV
from sklearn.linear_model import LinearRegression
```

```python
#Run Lasso Regression on engagement rate
```

```python
alpha = [0.0001, 0.0005, 0.001,0.003, 0.005,0.008, 0.01, 0.05, 0.1, 0.5, 1, 5, 10,50]
```

```python
#Engagement rate has high outliers, take the log to normalize
```

```python
DF1 = DF1.replace([np.inf, -np.inf],np.nan)
DF1 = DF1.dropna(axis=0)
```

```python
y = np.log(DF1['engagement_rate'])
x = DF1.loc[:, DF1.columns !='engagement_rate']
colnames = x.columns
```

```python
reg=LassoCV(alphas=alpha, cv=3, max_iter=1000, tol=0.0001).fit(x,y)
```

```python
reg.score(x,y)
```

```python
reg.alpha_
```

```python
reg_coef = reg.coef_ 
```

```python
print(sorted(zip(map(lambda x: round(x,4), reg_coef), colnames),reverse = True))
```
